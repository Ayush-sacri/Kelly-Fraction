```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
def monte_carlo_simulation(num_simulations, win_probability, reward_to_risk, initial_capital, risk_percent):
    final_capitals = []

    for _ in range(num_simulations):
        capital = initial_capital
        trades = np.random.choice([1, -1], size=num_trades, p=[win_probability, 1 - win_probability])

        for trade in trades:
                # Calculate position size based on risk percent
                position_size = capital * risk_percent

                # Calculate profit/loss for the trade
                pnl = position_size * trade * reward_to_risk

                # Update capital after the trade
                capital += pnl

        final_capitals.append(capital)

    return final_capitals

# Parameters
num_simulations = 1000
num_trades = 10
win_probability = 0.4
reward_to_risk = 2
initial_capital = 100000  # Initial capital
risk_percent = 0.02  # Risk 2% of capital per trade

# Run Monte Carlo simulation
final_capitals = monte_carlo_simulation(num_simulations, win_probability, reward_to_risk, initial_capital, risk_percent)

# Plot results


```
